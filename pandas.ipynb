```python
# Pandas is a powerful library in Python for data manipulation and analysis. 
# It provides data structures like Series and DataFrame, which allow you to 
# work with structured data efficiently. With Pandas, you can easily handle 
# missing data, perform operations on data, and visualize it. It is widely 
# used in data science and machine learning for tasks such as data cleaning, 
# exploration, and preprocessing.
```

```python
import pandas as pd
import numpy as np

# Dataframe is a 2-dimensional labeled data structure in pandas. 
# It is similar to a table in a relational database or an Excel spreadsheet. 
# A DataFrame consists of rows and columns, where each column can have a 
# different data type. It provides various functionalities for data 
# manipulation, such as filtering, grouping, and aggregation. DataFrames are 
# widely used in data analysis and machine learning tasks for handling and 
# processing structured data.


df = pd.DataFrame(np.arange(0,12).reshape(3,4) , index=['A','B','C'] , columns=['W','X','Y','Z'])
df.head()


#df.to_csv('mydata.csv')
#df.to_json('mydata.json')

df.loc['A']
# .loc[] is used to retrieve data from a particular row , but in case of .iloc[]
# it is used to retrieve data from a particular index position.
# index location.

df.iloc[0]
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(0,12).reshape(3,4) , index = ['1','2','3'] , columns = ['W','X','Y','Z'])
# First parameter is the data, second parameter is the index and third parameter is the columns.
# index is the label for the rows and it is used to identify the rows in the dataframe. 
# columns is the label for the columns.
df.to_excel('mydata.xlsx')


# the purpose of df.head() is display first few rows of the dataframe as we dont need to display the 
# entire dataframe , default is 5 rows.

df.loc['1']
type(df.iloc[1])
# type of df.iloc[1] is pandas.core.series.Series because it returns a single row as a Series object.
# series is a one-dimensional labeled array capable of holding any data type. It is similar to a column or a row in a DataFrame. 
# A Series has an index and a data value for each index. It can be created from a list, array, or dictionary. 
# Series are used for data manipulation and analysis in pandas.

print(np.array([df.iloc[1],df.iloc[2]]))

```
